```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(1,60,100)
r = 100
f_exp = np.exp(-np.square(a)/(2*np.square(r)))
f_bi = np.square(1-(np.square(a)/np.square(r)))
R= a.argsort().argsort()
f_r = np.exp(-R/r)

plt.clf() #close current figure
plt.plot(a,f_bi,label="bi")
plt.plot(a,f_exp,label="exp")
plt.plot(a,f_r,label="Ranking")
plt.legend()



```
